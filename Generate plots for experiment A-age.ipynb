```python
df = pd.read_csv('point_008_dpf.csv')
fig, ax = plt.subplots()
plt.bar(df["Age"],df["Actual"],yerr=[df["Actual"]-df["LCI"],df["HCI"]-df["Actual"]],color="green")
plt.xticks(rotation=90)
plt.xlabel("Age (dpf)")
plt.ylabel("Percentage that survived")
plt.title("0.008% MS-222 (% recovery)",fontweight="bold")
plt.tight_layout()
plt.savefig('point008_age')
plt.savefig('point008_age.eps')
```
